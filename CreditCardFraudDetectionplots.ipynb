```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import IsolationForest
from sklearn.neighbors import LocalOutlierFactor
from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score

# Step 1: Loading dataset
creditcardfraud = pd.read_csv('creditcard.csv')
```

```python
fig, axs = plt.subplots(ncols=2,nrows=3 ,figsize=(12,12))

class_counts = creditcardfraud['Class'].value_counts()
axs[0,0].pie(class_counts, labels=['Non-fraudulent (Class 0)', 'Fraudulent (Class 1)'], autopct='%1.1f%%', colors=['lightpink', 'blue'])
axs[0,0].set_title('Plot 1: Distribution of Class Variable')

sns.histplot(creditcardfraud['Time'], bins=50, kde=True, color='blue', ax=axs[0, 1])
axs[0, 1].set_title('Plot 2: Distribution of Transaction Time')
axs[0, 1].set_xlabel('Time (Seconds)')
axs[0, 1].set_ylabel('Frequency')

fraud_transactions = creditcardfraud[creditcardfraud['Class'] == 1]
non_fraud_transactions = creditcardfraud[creditcardfraud['Class'] == 0]

# Plot for Fraud Transactions
sns.histplot(fraud_transactions['Time'], bins=50, kde=True, color='teal', ax=axs[1, 0])
axs[1, 0].set_title('Plot 3: Distribution of Time in Fraudulent Transactions')
axs[1, 0].set_xlabel('Time (Seconds)')
axs[1, 0].set_ylabel('Frequency')

# Plot for Non-Fraud Transactions
sns.histplot(non_fraud_transactions['Time'], bins=50, kde=True, color='lightcoral', ax=axs[1, 1])
axs[1, 1].set_title('Plot 4: Distribution of Time in Non-Fraudulent Transactions')
axs[1, 1].set_xlabel('Time (Seconds)')
axs[1, 1].set_ylabel('Frequency')

# Plot for Fraud Transactions
sns.histplot(fraud_transactions['Amount'], color='teal', bins=50, ax=axs[2, 0])
axs[2, 0].set_title('Plot 5: Distribution of Amount in Fraudulent Transactions')
axs[2, 0].set_xlabel('Amount')
axs[2, 0].set_ylabel('Fequency')

# Plot for Non-Fraud Transactions
sns.histplot(non_fraud_transactions['Amount'], color='lightcoral', bins=50, ax=axs[2, 1])
axs[2, 1].set_title('Plot 6: Distribution of Amount in Non-Fraudulent Transactions')
axs[2, 1].set_xlabel('Amount')
axs[2, 1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```
